```python
import pandas as pd

#Create the data
characters_data = {
    'tv_characters_name': ['Goku', 'Luffy', 'Zoro', 'Usopp', 'Sanji', 'Nami', 'Brook', 'Robin', 'Franky', 'Jinbe'],
    'favorite_food': ['Eggs', 'Meat', 'Sushi', 'Tacos', 'French Toast', 'Tangerines', 'Milk', 'Sandwiches', 'Pizza', 'Donuts' ],
    'favorite_colors': ['Orange', 'Red', 'Green', 'Brown', 'Yellow', 'Gold', 'Black', 'Purple', 'White', 'Blue'],
    'birthplace': ['Japan', 'East Blue', 'East Blue', 'East Blue', 'East Blue', 'East Blue', 'East Blue', 'East Blue', 'East Blue', 'East Blue'],
    'tv_show_name': ['Dragon Ball', 'One Piece', 'One Piece', 'One Piece', 'One Piece', 'One Piece', 'One Piece', 'One Piece', 'One Piece', 'One Piece'],
    'ratings': [6.2, 8.4, 7.2, 8.7, 9.3, 6.5, 9.4, 7.9, 9.5, 6.5]
}

characters = pd.DataFrame(characters_data)
sort_ratings = characters.sort_values(by='ratings', ascending=False)

print(sort_ratings)
```

```python
import pandas as pd

#Create the data
characters_data = {
    'tv_characters_name': ['Goku', 'Luffy', 'Zoro', 'Usopp', 'Sanji', 'Nami', 'Brook', 'Robin', 'Franky', 'Jinbe'],
    'favorite_food': ['Eggs', 'Meat', 'Sushi', 'Tacos', 'French Toast', 'Tangerines', 'Milk', 'Sandwiches', 'Pizza', 'Donuts' ],
    'favorite_colors': ['Orange', 'Red', 'Green', 'Brown', 'Yellow', 'Gold', 'Black', 'Purple', 'White', 'Blue'],
    'birthplace': ['Japan', 'East Blue', 'East Blue', 'East Blue', 'East Blue', 'East Blue', 'East Blue', 'East Blue', 'East Blue', 'East Blue'],
    'tv_show_name': ['Dragon Ball', 'One Piece', 'One Piece', 'One Piece', 'One Piece', 'One Piece', 'One Piece', 'One Piece', 'One Piece', 'One Piece'],
    'ratings': [6.2, 8.4, 7.2, 8.7, 9.3, 6.5, 9.4, 7.9, 9.5, 6.5]
}

df = pd.DataFrame(characters_data)
filtered_df = df[(df['ratings'] > 8.0) & (df['tv_show_name'] == 'One Piece')]

print(filtered_df)
```
